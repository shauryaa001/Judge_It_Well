(ii) Using loops (without using numpy)

```python
#The inputs here are normal 2D arrays

def Validmult(m1, m2):
  if len(m1[0]) == len(m2):
    return True
  return False

def multiplymatrices(m1, m2):
  if not Validmult(m1, m2):
    print("ERROR: Given Multiplication is not valid")
    return None
  res = []
  for i in range(len(m1)):
    a = []
    for j in range(len(m2[0])):
      element = 0
      for k in range(len(m1[0])):
        element += m1[i][k]*m2[k][j]
      a.append(element)
    res.append(a)
  return res

matc1 = [[numpy.random.randint(11,100) for i in range(4)] for j in range(5)]
matc2 = [[numpy.random.randint(11,100) for i in range(5)] for j in range(4)]
print(matc1)
print(matc2)
print(multiplymatrices(matc1, matc2))
```
